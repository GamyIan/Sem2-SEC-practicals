```python
import pandas as pd
import numpy as np
```

```python
# Loading data and displaying first 5 records
df = pd.read_csv("Netflix_Dataset.csv")
df.head()
```

```python
# How many unique TV shows and Movies
df["type"].value_counts()
```

```python
# What is total number of missing values in each column
df.isnull().sum()
```

```python
# what are the different genres available in the dataset
df["genre"].unique()
```

```python
# How many unique directors are there in the dataset?
df["director"].nunique()
```

```python
# Find all movies released in 2019.
df[(df['type']=="Movie") & (df['release_year']==2019)]
```

```python
# Retrieve all TV Shows directed by Neeraj Pandey.
df[(df['type']=="TV Show") & (df['director']=="Neeraj Pandey")]
```

```python
# List all the movies where Akshay Kumar is part of the cast.
df[(df['type']=="Movie") & (df['cast'].str.contains("Akshay Kumar"))]
```

```python
# Show all content that has a rating of UA and a duration of more than 150 minutes.
UA_movies=df[(df['type']=="Movie") & (df['rating']=="UA")] 
UA_movies[UA_movies['duration'].str.replace("min","").astype(int)>150]
```

```python
# 10. Find all content where the genre is either Comedy or Drama.
df[(df['genre']=="Comedy") | (df['genre']=="Drama")]
```

```python
# 11. Sort the dataset by release year in descending order.
df.sort_values("release_year",ascending=False).head()
```

```python
# 12. Sort the dataset first by type and then by title alphabetically.
df.sort_values(["type","title"])#,ascending=[True,True])
```

```python
# 13. Find the number of movies and TV shows released per year.
df.groupby("release_year").count().iloc[:,:1]
```

```python
# 14. What are the top 3 most common genres and their counts?
df.groupby("genre").count().iloc[:,:1].sort_values("show_id",ascending=False)[:3]
```

```python
# 15. Find the number of movies directed by each director.
df.value_counts("director")
```

```python
# 16. Group the dataset by rating and count how many titles each rating has.
df.value_counts("rating")
```
